```python
# Deal with the missing values of Type-1 columns: Price, Season, NeckLine, SleeveLength, Winter and Autumn.

# List of Type-1 columns
type1_cols = ['Price', 'Season', 'NeckLine', 'SleeveLength', 'Winter', 'Autumn']

# Strategy 1: Fill with mode for categorical columns
categorical_cols = ['Season', 'NeckLine', 'SleeveLength','Price']
for col in categorical_cols:
    mode_value = inp0[col].mode()[0]
    print(mode_value)
    inp0[col] = inp0[col].fillna(mode_value)

# Strategy 2: Fill with median for numerical columns
numerical_cols = ['Winter', 'Autumn']
for col in numerical_cols:
    median_value = inp0[col].median()
    print(median_value)
    inp0[col] = inp0[col].fillna(median_value)

# Verify no missing values remain in Type-1 columns
print("Remaining missing values in Type-1 columns:")
print(inp0[type1_cols].isnull().sum())
#mode and median value that has updated to the left
```
