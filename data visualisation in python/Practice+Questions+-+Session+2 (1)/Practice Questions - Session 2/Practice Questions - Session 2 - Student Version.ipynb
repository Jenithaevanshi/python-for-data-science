## 1. Average
Q1)The absolute difference in the average profit percentages of Delhi and Mumbai comes out to be approximately ____

- a) 1.67

- b) 1.57

- c) 1.77

- d) 1.47


```python
# Calculate average profit percentage for each city's column
import pandas as pd

# Load the dataset
df = pd.read_csv("Footwear_v2.csv")

# Clean percentage columns and convert to numeric
def clean_percentage(column):
    return (
        column.str.replace('%', '', regex=False)  # Remove % sign
        .astype(float)                           # Convert to float
    )

df['Delhi'] = clean_percentage(df['Delhi'])
df['Mumbai'] = clean_percentage(df['Mumbai'])

# Calculate averages
avg_delhi = df['Delhi'].mean()
avg_mumbai = df['Mumbai'].mean()

# Compute absolute difference
abs_diff = abs(avg_delhi - avg_mumbai)

print(f"Average Profit Percentage - Delhi: {avg_delhi:.2f}%")
print(f"Average Profit Percentage - Mumbai: {avg_mumbai:.2f}%")
print(f"Absolute Difference: {abs_diff:.2f} percentage points")
```

## 2. Box Plots

Plot a box plot to analyse the spread of profits for each of the cities. Which city has the highest profit value at the upper fence in the box plot?



- a) Delhi

- b) Mumbai

- c) Hyderabad   ans:

- d) Jaipur


```python
#Your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Footwear_v2.csv")

# Clean percentage columns and convert to numeric
def clean_percentage(column):
    return (
        column.str.replace('%', '', regex=False)  # Remove % sign
        .astype(float)                           # Convert to float
    )

df['Delhi'] = clean_percentage(df['Delhi'])
df['Mumbai'] = clean_percentage(df['Mumbai'])
df['Hyderabad'] = clean_percentage(df['Hyderabad'])
df['Jaipur'] = clean_percentage(df['Jaipur'])
# Reshape data for boxplot (combine both cities into one column)
df_melted = pd.melt(df[['Delhi', 'Mumbai','Hyderabad','Jaipur']], var_name='City', value_name='Profit Percentage')

# Create boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='City', y='Profit Percentage', data=df_melted)
plt.title('Profit Percentage Distribution by City')
plt.ylabel('Profit Percentage (%)')
plt.xlabel('City')
plt.show()

# Calculate upper fences
def get_upper_fence(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return Q3 + (1.5 * IQR)

delhi_upper = get_upper_fence(df['Delhi'])
mumbai_upper = get_upper_fence(df['Mumbai'])
Hyderabad_upper = get_upper_fence(df['Hyderabad'])
Jaipur_upper = get_upper_fence(df['Jaipur'])                                  

```

A negative correlation on the other hand(values near -1) indicate good negative trend - if one increases, then the other decreases. A value near 0 indicates no correlation, as in one variable doesn’t affect the other.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (assuming it contains the mentioned cryptocurrencies)
df = pd.read_csv("crypto.csv")

# Select only numeric columns (assuming all price columns are numeric)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df_numeric = df[numeric_cols]

# Create correlation matrix
corr_matrix = df_numeric.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f",
            annot_kws={"size": 10}, linewidths=0.5)
plt.title('Cryptocurrency Price Correlations', pad=20, fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Evaluate statements
print("\nStatement Evaluation:")
# 1. Ethereum and Quantum have a high correlation
eth_quantum = corr_matrix.loc['ethereum', 'quantum']
print(f"Ethereum-Quantum correlation: {eth_quantum:.2f} → {'True' if eth_quantum > 0.7 else 'False'}")

# 2. Neo and Bitcoin have pretty low correlation
neo_bitcoin = corr_matrix.loc['neo', 'bitcoin']
print(f"Neo-Bitcoin correlation: {neo_bitcoin:.2f} → {'True' if abs(neo_bitcoin) < 0.3 else 'False'}")

# 3. Ethereum has similar correlation with Litecoin and Neo
eth_litecoin = corr_matrix.loc['ethereum', 'litecoin']
eth_neo = corr_matrix.loc['ethereum', 'neo']
similarity_threshold = 0.1  # Maximum allowed difference
print(f"Ethereum-Litecoin: {eth_litecoin:.2f}, Ethereum-Neo: {eth_neo:.2f} → "
      f"{'True' if abs(eth_litecoin - eth_neo) < similarity_threshold else 'False'}")
```
